# Como se podria representar una EDO en python


La representacion de una EDO puede ser simplemente usar la variable x como independiente y la funcion Y dependiente de x, donde Z como la derivada de de Y:

```python
X = 0
Y = 3*X
Z = 3*Y + X
```

Con esto definimos una "EDO" con un valor de x = 0, pero si queremos el campo de direcciones e isoclinas, se vuelve una tarea muy tediosa, ya que no tenemos forma de representarla graficamente que no sea la consola y ademas de estar limitado a la forma de obtencion de las direcciones por lo que en este curso se usara numpy, matplotlib (versiones por confirmar), para la representacion y obtencion de estos datos y precesarlo.
en esta ayudantia esto se compilo con numpy Version: 2.2.4 y matplotlib Version: 3.10.1
Los puedes instalar en la anaconda promp con: pip install numpy==2.2.4 matplotlib==3.10.1 (aseguras de tener la misma version)

```python
# primero que todo para usar numpy (operaciones matematicas) y pyplot de matplotlib (representacion grafica) debemos de importarlas
import numpy as np # as significa llamarlo de otro modo para evitar escribir mas
import matplotlib.pyplot as plt # ejemplo "import paralelepipedo as pipe" seria mas corto que escribir paralelepipedo
```

#### Con la ayuda de numpy y matplotlib podremos representar un campo de direcciones, la representacion que se espera en laboratorio

##### Sea la edo 
$$y'=-2y+2\cos(x)$$

```python
# definimos las variables a analizar y su intervalo a analizar con meshgrid y linspace de numpy
X,Y=np.meshgrid(np.linspace(-2,5,20), np.linspace(-10,10,12)) # Formato de linspace es (limite izquierdo, limite derecho, {opcional: cantidad de puntos})

U = 1.0
V = -2*Y+2*np.cos(X) #f(x,y), definimos la edo

N = np.sqrt(U ** 2 + V ** 2)# Normalizamos los vectores, ya que queremos saber los sentidos, no queremos los largos de los vectores
# al obtener la normalizacion se debe de dividir con los vectores, para asi terminar con la normalizacion
U = U / N
V = V / N

# invocamos plt para graficar los vectores
plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y trmino en la normalizacion (u,v)
plt.xlim([-2,5]) #limites en eje x
plt.ylim([-10,10]) #limites en eje y

plt.xlabel('x') #etiqueta eje x
plt.ylabel('y') #etiqueta eje y
plt.grid()# para establecer red en la regin

# mostramos el grafico
plt.show()
```

# Representacion de una funcion

## Considerando la EDO anterior

$$y'=-2y+2\cos(x)$$

## La cual cuya solucion general es

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}+\dfrac{C}{{e}^{2\,x}}$$

Entendiendo C como una constante, de la cual si entregamos valores a C obtendremos miembros de la familia de soluciones

para ejemplificar entregaremos 5 valores a C:

### Funcion 1 (C = 0):

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}$$

### Funcion 2 (C = 2):

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}+\dfrac{2}{{e}^{2\,x}}$$

### Funcion 3 (C = 3):

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}+\dfrac{3}{{e}^{2\,x}}$$

### Funcion 4 (C = 5):

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}+\dfrac{5}{{e}^{2\,x}}$$

### Funcion 5 (C = 200):

$$y=\dfrac{2\,\operatorname{sen}\left(x\right)}{5}+\dfrac{4\,\cos\left(x\right)}{5}+\dfrac{200}{{e}^{2\,x}}$$

# Que pasa si graficamos estas funciones?

```python
# definimos los limites
plt.xlim([-2, 5]) #limites en eje x
plt.ylim([-10, 10]) #limites en eje y

# etiquetas
plt.xlabel('x') #etiqueta eje x
plt.ylabel('y') #etiqueta eje y

# definimos x, ademas de sus intervalos
x=np.linspace(-2,5, 200)

# escribimos las funciones a graficar
y1 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5)
y2 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (2 * np.exp(-2 * x))
y3 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (3 * np.exp(-2 * x))
y4 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (5 * np.exp(-2 * x))
y5 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (200 * np.exp(-2 * x))

# estas opciones son para cambar caracteristicas de la funcion en el color formato de 0 a 1 (ausensia o intensidad respectivamente)
# OJO aqui y1 es una variable cualquiera, no tiene que ser y1 o y2, pero ya que normalmente se representa como y = f(x)
# recomiendo mucho esta representacion para mejor claridad a la hora de revisar o trabajar en esto
plt.plot(x,y1,color=(1,0.2,0.3))
plt.plot(x,y2,color=(0.1,0.2,0.3))
plt.plot(x,y3,color=(0.1,0.2,0.3))
plt.plot(x,y4,color=(0.1,0.2,0.3))
plt.plot(x,y5,color=(0.1,0.2,0.3))

# mostramos la grafica creada
plt.show()
```

# Y que pasa si insertamos estas funciones en el campo de direcciones?:

```python
# formato para crear un campo de direcciones:
X,Y=np.meshgrid(np.linspace(-5,5,20), np.linspace(-10,10,12)) # Formato de linspace es (limite izquierdo, limite derecho, {opcional: cantidad de puntos})

U = 1.0
V = -2*Y+2*np.cos(X) #f(x,y), definimos la edo

N = np.sqrt(U ** 2 + V ** 2)#Normalizamos los vectores, ya que queremos saber los sentidos, no queremos los largos de los vectores
# al obtener la normalizacion se debe de dividir con los vectores, para asi terminar con la normalizacion
U = U / N
V = V / N

# invocamos plt para graficar los vectores
plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y trmino en la normalizacion (u,v)
plt.xlim([-2,5]) #limites en eje x
plt.ylim([-10,10]) #limites en eje y


# ahora invocamos para ingresar funciones
x=np.linspace(-2,5, 200)

# escribimos las funciones a graficar
y1 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5)
y2 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (2 * np.exp(-2 * x))
y3 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (3 * np.exp(-2 * x))
y4 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (5 * np.exp(-2 * x))
y5 = (2 * np.sin(x) / 5) + (4 * np.cos(x) / 5) + (200 * np.exp(-2 * x))

# estas opciones son para cambar caracteristicas de la funcion en el color formato de 0 a 1 (ausensia o intensidad respectivamente)
plt.plot(x,y1,color=(1,0.2,0.3))
plt.plot(x,y2,color=(0.1,0.2,0.3))
plt.plot(x,y3,color=(0.1,0.2,0.3))
plt.plot(x,y4,color=(0.1,0.2,0.3))
plt.plot(x,y5,color=(0.1,0.2,0.3))

# mostramos la grafica creada
plt.show()
```

# Ejemplo 1

suponga las siguientes EDO:

* $y'=x-y-5$
* $y'=\sin(x)-y$
* $y'=\tan(x)+y$
* $y'={e}^{x}$
* $y'=ey-x$


- a) obtener el campo de direcciones entre los rangos $[-4,10] \times [-10,10]$
- b) Grafica miembros de la familia de soluciones de las EDO e insertalos en el campo de direcciones

```python
# definimos las variables a analizar y su intervalo a analizar con meshgrid y linspace de numpy
X,Y=np.meshgrid(np.linspace(-4,10,20), np.linspace(-10,10,12)) # Formato de linspace es (limite izquierdo, limite derecho, {opcional: cantidad de puntos})

U = 1.0
V = X - Y - 5 #f(x,y), definimos la edo

N = np.sqrt(U ** 2 + V ** 2)# Normalizamos los vectores, ya que queremos saber los sentidos, no queremos los largos de los vectores
# al obtener la normalizacion se debe de dividir con los vectores, para asi terminar con la normalizacion
U = U / N
V = V / N

# invocamos plt para graficar los vectores
plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y trmino en la normalizacion (u,v)
plt.xlim([-4,10]) #limites en eje x
plt.ylim([-10,10]) #limites en eje y

plt.xlabel('x') #etiqueta eje x
plt.ylabel('y') #etiqueta eje y
plt.grid()# para establecer red en la regin

# mostramos el grafico
plt.show()
```
